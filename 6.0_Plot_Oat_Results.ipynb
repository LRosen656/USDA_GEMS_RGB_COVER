```python
##Defining RMSE
def rmse(x,y, summary):                                 
    y_true = y
    slope = summary.slope
    intercept = summary.intercept
    y_pred= x*slope + intercept                       ##Line Equation
    result = np.sqrt(((y_pred - y_true)**2).mean())
    return(result)
```
