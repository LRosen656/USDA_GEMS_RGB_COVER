```python
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage.io import imread, imsave
import glob
```

```python
src = ("L:/USDA/images_crop_thirdoffset_tif/")
exg_dir =  "L:/USDA/EXG"
exgr_dir = "L:/USDA/EXGR"
gli_dir = "L:/USDA/GLI"
vari_dir = "L:/USDA/VARI"
if not os.path.exists(exg_dir):
    os.mkdir(exg_dir)
if not os.path.exists(exgr_dir):
    os.mkdir(exgr_dir)
if not os.path.exists(gli_dir):
    os.mkdir(gli_dir)
if not os.path.exists(vari_dir):
    os.mkdir(vari_dir)

os.chdir(src)
    
```

```python
def EXG(output = None, Red_N= None, Green_N= None, Blue_N=None):
    os.chdir(exg_dir)
    result = (2.0 * Green_N) - Red_N - Blue_N
    imsave(output + "_EXG.tif", result)
```

```python
def EXGR(output = None, Red_N= None, Green_N= None, Blue_N=None):
    os.chdir(exgr_dir)
    result = (3.0 * Green_N) - (2.4 * Red_N) - Blue_N
    imsave(output + "_EXGR.tif", result)
```

```python
def GLI(output = None, Red_C= None, Green_C= None, Blue_C=None):
    os.chdir(gli_dir)
    result = ((2.0 * Green_C) - Red_C - Blue_C)/((2.0 * Green_C) + Red_C + Blue_C)
    result[np.isnan(result)] = 0
    result[np.isnan(result)] = 0
    imsave(output + "_GLI.tif", result)
```

```python
def GLI(output = None, Red_C= None, Green_C= None, Blue_C=None):
    os.chdir(gli_dir)
    result = ((2.0 * Green_C) - Red_C - Blue_C)/((2.0 * Green_C) + Red_C + Blue_C)
    result[np.isnan(result)] = 0
    result[np.isnan(result)] = 0
    imsave(output + "_GLI.tif", result)
```

```python
def VARI(output = None, Red_C= None, Green_C= None, Blue_C=None):
    os.chdir(vari_dir)
    result = ((1.0 * Green_C) - Red_C)/((1.0 * Green_C) + Red_C + Blue_C)
    result[np.isnan(result)] = 0
    result[np.isnan(result)] = 0
    imsave(output + "_VARI.tif", result)
```

```python
with np.errstate(divide = "ignore", invalid = "ignore"):
    print("Starting Vegetation Indices...")
    for tiff in glob.glob("*.tif"):
        os.chdir(src)
        name = tiff.split(".")[0]
        
        img = imread(tiff)
        
        Red_c = img[:,:,0]

        Green_c = img[:,:,1]

        Blue_c = img[:,:,2]


        Red = Red_c/Red_c.max()

        Green = Green_c/Green_c.max()

        Blue = Blue_c/Blue_c.max()

        Red_n = Red/(Red + Green + Blue)
        Red_n[np.isnan(Red_n)] = 0
        Red_n[np.isinf(Red_n)] = 0

        Green_n = Green/(Red + Green + Blue)
        Green_n[np.isnan(Green_n)] = 0
        Green_n[np.isinf(Green_n)] = 0

        Blue_n = Blue/(Red + Green + Blue)
        Blue_n[np.isnan(Blue_n)] = 0
        Blue_n[np.isinf(Blue_n)] = 0
        
        print("Created color matrices...\n")
        
        EXG(name, Red_n, Green_n, Blue_n)
        EXGR(name, Red_n, Green_n, Blue_n)
        GLI(name, Red_c, Green_c, Blue_c)
        VARI(name, Red_c, Green_c, Blue_c)
        print("Done with image " + name + "...\n")
        os.chdir(src)
    print("All images done...\n")
print("Vegetaion Index Complete!!!")
```
