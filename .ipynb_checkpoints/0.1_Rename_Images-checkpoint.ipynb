# Renaming USDA Oat Images

This notebook renames the images taken from an oat study over the summer of 2021. A for loop was used for each date directory to copy the image to a new date directory and labeled with the correct plot number and row. Date 7/22 was skipped because it just had soybean images. Date 7/8 was already labeled so it just needed to be copied.

Images formats are changed to tiffs. 

```python
import os
import pandas as pd
import glob
import shutil
```

```python
src_dir = "~/images/" ##Directory with all the image directories.
if not os.path.exists("~/images_tif/"):
    os.mkdir("L:/USDA/images_tif/")
new_dir = "L:/USDA/images_tif/"     ##Directory the images will be copied. 
os.chdir(src_dir)
Plot_Order = pd.read_csv("Plot_Order - Picture Order.csv") ##Spreadsheet of the plot/row order
for folder in glob.glob('*'): ##Everything in the current directiory
    os.chdir(src_dir)
    if folder == 'Plot_Order - Picture Order.csv': ##Skipping the spreadsheet
        print("Skipping the csv...\n")
    elif folder == "7.22.2021":                    ##Skipping Date with just soybeans  
        print("Skipping folder " + folder + "...\n")
    else:
        if not os.path.exists(new_dir + folder): 
            os.mkdir(new_dir + folder)
            print("Directory " + new_dir + folder + " is created...\n")
        if folder == "7.8.2021":                                         ##Date 7/8 is already labeled so I just need to copy it. 
            print("Working in images in folder " + folder + "...\n")
            os.chdir(src_dir + folder)
            for jpeg in sorted(glob.glob("*.jpg")):
                name = jpeg.split(".") ## Splitting the jpeg
                print("Copying " + jpeg + "...\n")
                shutil.copyfile(jpeg, new_dir + folder + "/" + name[0] + ".tif")
            print("Done with folder " + folder + "...\n")
        else:
            print("Working on images in " + folder + "...\n")
            os.chdir(src_dir + folder)
            for (jpeg, plot, row) in zip(sorted(glob.glob("*.jpg")), Plot_Order["plot"], Plot_Order["row"]):
                shutil.copyfile(jpeg, new_dir + folder + "/" + str(plot) + str(row.capitalize()) + ".tif") ##Copies image with correct name to new directory
                print("Copied jpeg " + str(plot) + str(row.capitalize()) + "...\n")                       
    print("Done with " + folder + "...\n")
print("All Folders done...\n")
print("Renaming images Complete!!!")

```
